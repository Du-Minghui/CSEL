```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("activity.csv")
```

```python
df
```

```python
df["activity"] = df["activity"].str.split(";")
```

```python
df2 = df.explode("activity")
```

```python
df2
```

```python
counts = df2[['BGC_ID', 'activity']].pivot_table(index='BGC_ID', 
                                                           columns='activity',
                                                           aggfunc=np.count_nonzero).fillna(0)
```

```python
counts
```

```python
counts2 = counts.astype(bool).astype(int)
```

```python
counts2.to_csv("antibacterial.csv")
```
