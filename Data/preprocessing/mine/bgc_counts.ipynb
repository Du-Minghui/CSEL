```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("mibig_bgcs_all.csv")
```

```python
df
```

```python
counts = df[['contig_id', 'pfam_id']].pivot_table(index='contig_id', 
                                                           columns='pfam_id',
                                                           aggfunc=np.count_nonzero).fillna(0)
```

```python
counts2 = counts.astype(bool).astype(int)
```

```python
counts2
```

```python
counts2.to_csv("bgc_counts.csv")
```
