```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("bgc_pfam.csv")
```

```python
df["pfam_ids"] = df["pfam_ids"].str.split(" ")
```

```python
df2 = df.explode("pfam_ids")
```

```python
df2
```

```python
df2["pfam_ids"] = df2['pfam_ids'].str.split('.',expand=True)[0]
```

```python
df2
```

```python
counts = df2[['bgc_ids', 'pfam_ids']].pivot_table(index='bgc_ids', 
                                                  columns='pfam_ids',
                                                  aggfunc=np.count_nonzero).fillna(0)
counts = counts.astype(bool).astype(int)
```

```python
counts
```

```python
df1= pd.read_csv("mibig(3.1).csv")
```

```python
df1
```

```python
def read_vectors_to_predict(vectors, train_columns):
    for col in set(train_columns).difference(vectors.columns):
        vectors[col] = 0
    return vectors[train_columns]
```

```python
columns = df1.columns.tolist()
```
