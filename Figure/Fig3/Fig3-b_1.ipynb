```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Arial"
plt.rcParams['axes.unicode_minus'] = False  
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

```python
genus_gcf = pd.read_csv("gcf_genus.csv",index_col=0)
```

```python
genus_gcf
```

```python
unit_topics = genus_gcf.columns.tolist()
```

```python
labels = genus_gcf.index
```

```python
a = genus_gcf.iloc[0].tolist()
b = genus_gcf.iloc[1].tolist()
c = genus_gcf.iloc[2].tolist()
d = genus_gcf.iloc[3].tolist()
e = genus_gcf.iloc[4].tolist()
f = genus_gcf.iloc[5].tolist()
g = genus_gcf.iloc[6].tolist()
h = genus_gcf.iloc[7].tolist()
i = genus_gcf.iloc[8].tolist()
j = genus_gcf.iloc[9].tolist()
k = genus_gcf.iloc[10].tolist()
l = genus_gcf.iloc[11].tolist()
```

```python
c_bottom = np.add(a, b)
d_bottom = np.add(c_bottom, c)
e_bottom = np.add(d_bottom, d)
f_bottom = np.add(e_bottom, e)
g_bottom = np.add(f_bottom, f)
h_bottom = np.add(g_bottom, g)
i_bottom = np.add(h_bottom, h)
j_bottom = np.add(i_bottom, i)
k_bottom = np.add(j_bottom, j)
l_bottom = np.add(k_bottom, k)
```

```python
bar_width = 0.5

plt.figure(figsize=(10, 8))

plt.ylim(bottom=0,top=180)

plt.bar(range(len(a)), a, bar_width ,color="#e96162")
plt.bar(range(len(b)), b,bar_width , bottom = a, color="#fab166")
plt.bar(range(len(c_bottom)), c, bar_width , bottom = c_bottom, color="#ffd669")
plt.bar(range(len(d_bottom)), d, bar_width , bottom = d_bottom, color="#8dc781")
plt.bar(range(len(e_bottom)), e, bar_width , bottom = e_bottom, color="#77a4b3")
plt.bar(range(len(f_bottom)), f, bar_width , bottom = f_bottom, color="#6ba9e0")
plt.bar(range(len(g_bottom)), g, bar_width , bottom = g_bottom, color="#c775a8")
plt.bar(range(len(h_bottom)), h, bar_width , bottom = h_bottom, color="#ffff00")
plt.bar(range(len(i_bottom)), i, bar_width , bottom = i_bottom, color="#e33199")
plt.bar(range(len(j_bottom)), j, bar_width , bottom = j_bottom, color="#bcbdb9")
plt.bar(range(len(k_bottom)), k, bar_width , bottom = k_bottom, color="#741b47")
plt.bar(range(len(l_bottom)), l, bar_width , bottom = l_bottom, color="#2a4e71")



ax = plt.subplot()
ax.set_xticks(range(len(unit_topics)))


plt.legend(labels,fontsize=13,framealpha=1,fancybox=False,edgecolor='black')

ax.set_ylim(bottom=0,top=150)
ax.tick_params(labelsize=15)

ax.set_xlabel('',size=20,labelpad=15)
ax.set_ylabel('',size=20,labelpad=15)




ax.set_xticklabels(unit_topics)
#ax.set_title("Bacterial genus distribution of GCFs")
ax.set_xlabel("Natural products class")
ax.set_ylabel("GCF counts")

plt.tight_layout()

plt.savefig('Fig3-b_1.pdf', bbox_inches='tight')
plt.show()
```
