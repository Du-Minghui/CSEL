```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Arial"
plt.rcParams['axes.unicode_minus'] = False  
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['axes.unicode_minus'] = False 
from matplotlib import cm
```

```python
data = pd.read_csv("class_genus.csv")
```

```python
data
```

```python
sizes_inner = data["BGC_num1"][:5,]
sizes_outer = data["BGC_num2"]
```

```python
colors_outer = data['colors_outer3']
colors_inner = data['colors_inner'][:5,]
```

```python
plt.subplots(figsize=(10,10))

plt.pie(sizes_outer, colors=colors_outer,radius=0.9,autopct='%0.2f%%',pctdistance=1.1,
        textprops={'fontsize': 15, 'color': 'k'},
        )



plt.pie(sizes_inner, colors=colors_inner,radius=0.6,autopct='%0.2f%%',pctdistance=0.8,labeldistance=0.3,
        labels=["NRPS","Ripps","PKS","Terpene","Saccharides"], textprops={'fontsize': 15, 'color': 'k'},
        wedgeprops=dict(width=0.1, edgecolor='w'))

circle=plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(circle)

#plt.title('class_genus_bgc')



plt.tight_layout()

plt.savefig('Fig3-b_2.pdf', bbox_inches='tight')
plt.show()
```
