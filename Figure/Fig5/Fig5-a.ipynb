```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
import palettable 
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Arial"
plt.rcParams['axes.unicode_minus'] = False  
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['axes.unicode_minus'] = False  
```

```python
data = pd.read_csv("core_class.csv")
```

```python
data
```

```python
sizes = data["count"]
```

```python
label = data['Core Peptide']
```

```python
colors=palettable.cartocolors.qualitative.Pastel_10.mpl_colors


plt.subplots(figsize=(8,8))

plt.pie(sizes,colors=colors, autopct='%.2f%%',pctdistance=1.2,radius=0.7, textprops={'fontsize': 15, 'color': 'k'},
        wedgeprops=dict(width=0.2, edgecolor='w'))

circle=plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(circle)

plt.title('Core class(BAGEL4)',fontsize=14,x=0.5,y=0.47)

plt.legend(label,bbox_to_anchor=(1.25, 0.5),loc='center right',fontsize=12,framealpha=1,fancybox=False,edgecolor='black')


#plt.tight_layout()

plt.savefig('Fig5-a.pdf', bbox_inches='tight')
plt.show()
```
