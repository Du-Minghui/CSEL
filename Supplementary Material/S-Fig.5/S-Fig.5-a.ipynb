```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
import palettable 
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.unicode_minus'] = False  
```

```python
data = pd.read_csv("pie_napdos_nodup.csv")
```

```python
data
```

```python
sizes = data["num"]
```

```python
label = data['BGC match']
```

```python
colors=palettable.cartocolors.qualitative.Pastel_10.mpl_colors


plt.subplots(figsize=(8,8))

plt.pie(sizes,colors=colors, autopct='%.2f%%',pctdistance=1.2,radius=0.7, textprops={'fontsize': 15, 'color': 'k'},
        wedgeprops=dict(width=0.2, edgecolor='w'))

circle=plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(circle)

plt.title('C domains(NaPDoS2)',fontsize=16,x=0.5,y=0.47)

plt.legend(label,bbox_to_anchor=(1.4, 0.5),loc='center right',fontsize=12,framealpha=1,fancybox=False,edgecolor='black')


plt.tight_layout()

plt.savefig('Fig5-A.pdf', bbox_inches='tight')
plt.show()
```
