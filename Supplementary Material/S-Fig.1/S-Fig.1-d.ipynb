```python
import numpy as np
import pandas as pd
import math
import srd
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.unicode_minus'] = False  
```

```python
df1 = p_m = pd.read_csv('p_metric_29.csv',index_col=0)
df2 = n_m = pd.read_csv('n_metric_29.csv',index_col=0)
```

```python
ref1 = 'max'
ref2 = 'min'
```

```python
srd_collector1=[]
for ref1 in df1.columns:
    srd_current=srd.srd_core(df1,ref1)
    srd_collector1.append(srd_current)

srd_collector1=pd.DataFrame(srd_collector1)
srd_collector1 = srd_collector1.iloc[::-1]
srd_collector1_row = srd_collector1.columns.to_list()[::-1]
srd_collector1_columns = srd_collector1.columns.to_list()
```

```python
gs = plt.GridSpec(1, 2, width_ratios=(9, 1))

fig, ax = plt.subplots(figsize=(10,10))

ax = plt.subplot(gs[0,0])

cmap=plt.get_cmap('YlGnBu')

pc = plt.imshow(srd_collector1, cmap = cmap)

ax.set_xticks(np.arange(len(srd_collector1)))
ax.set_yticks(np.arange(len(srd_collector1)))
ax.set_xticklabels(srd_collector1_columns)
ax.set_yticklabels(srd_collector1_row)

plt.xticks(rotation=90)
ax.tick_params(labelsize=15)

ax = plt.subplot(gs[0,1])
plt.colorbar(pc, cax=ax)
plt.title('SRD (%)')

fig.tight_layout()
plt.savefig('Fig2-E1.pdf', bbox_inches='tight')
plt.show()
```

```python
srd_collector2=[]
for ref2 in df2.columns:
    srd_current=srd.srd_core(df2,ref2)
    srd_collector2.append(srd_current)

srd_collector2 = pd.DataFrame(srd_collector2)

srd_collector2.index = srd_collector2.columns

srd_collector2 = srd_collector2.iloc[::-1]
srd_collector2_row = srd_collector2.columns.to_list()[::-1]
srd_collector2_columns = srd_collector2.columns.to_list()
```

```python
gs = plt.GridSpec(1, 2, width_ratios=(9, 1))

fig, ax = plt.subplots(figsize=(10,10))

ax = plt.subplot(gs[0,0])

cmap=plt.get_cmap('YlGnBu')

pc = plt.imshow(srd_collector2, cmap = cmap)

ax.set_xticks(np.arange(len(srd_collector2)))
ax.set_yticks(np.arange(len(srd_collector2)))
ax.set_xticklabels(srd_collector2_columns)
ax.set_yticklabels(srd_collector2_row)

plt.xticks(rotation=90)
ax.tick_params(labelsize=15)

ax = plt.subplot(gs[0,1])
plt.colorbar(pc, cax=ax)
plt.title('SRD (%)')

fig.tight_layout()
plt.savefig('Fig2-E2.pdf', bbox_inches='tight')
plt.show()
```
