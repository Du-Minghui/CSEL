```python
import pandas as pd
import numpy as np
from matplotlib import font_manager as fm
from  matplotlib import cm
import palettable 
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.unicode_minus'] = False  
```

```python
data = pd.read_csv("total.csv")
```

```python
sizes = values = data["counts"]
labels = data["part"]
explode = data["explode"]
```

```python
colors = ["#faee66", "#faee66", "#01f9c6", "#01f9c6"]
plt.figure(figsize=(8,8))
patches,text,p_text = plt.pie(sizes,explode=explode, autopct='%0.2f%%',colors=colors,pctdistance=1.3)
patches[1].set_hatch('/') 
patches[1].set_edgecolor('#DC143C')
patches[1].set_linewidth(1.5)
patches[1].set_linestyle('--') 
patches[2].set_hatch('/') 
patches[2].set_edgecolor('#DC143C')
patches[2].set_linewidth(1.5)
patches[2].set_linestyle('--') 
circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)

for t in p_text:
    t.set_size(25)

plt.title('Total',x=0.5,y=0.45,fontsize=45)
plt.savefig('Fig1.D1.pdf', bbox_inches='tight')
plt.show()
```
