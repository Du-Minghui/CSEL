```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.unicode_minus'] = False  
```

```python
from matplotlib import font_manager as fm
from  matplotlib import cm
```

```python
import palettable 
```

```python
NPAtlas = pd.read_csv("NPAtlas_Bacterium.csv")
```

```python
NPAtlas_sizes =values = NPAtlas["counts"]
NPAtlas_labels = NPAtlas["genus"][0:20]
NPAtlas_explode = NPAtlas["explode"]
```

```python
NPAtlas_mian_labels = ['{} {:1.2f}% ({})'.format(i, j, v) for i, j, v in
                       zip(NPAtlas_labels, 100. * values / values.sum(), values)]
```

```python
NPAtlas_mian_labels
```

```python
NPAtlas_map_vir = cm.get_cmap(name='PuBu')
NPAtlas_colors = NPAtlas_map_vir(NPAtlas_sizes)
plt.figure(figsize=(8,8))
plt.pie(NPAtlas_sizes,explode=NPAtlas_explode,colors=NPAtlas_colors)
circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
plt.legend(NPAtlas_mian_labels,loc='center right',bbox_to_anchor=(1.7, 0.5), fontsize=17,title='Top 20',
           framealpha=1,fancybox=False,edgecolor='black')
plt.title('NPAtlas',fontsize=35,x=0.5,y=0.47) 
plt.savefig('Fig1.A.pdf', bbox_inches='tight')
plt.show()
```

```python
NCBI = pd.read_csv("NCBI_Bacterium.csv")
```

```python
NCBI_sizes = values = NCBI["counts"]
NCBI_labels = NCBI["genus"]
NCBI_explode = NCBI["explode"]
```

```python
NCBI_main_labels = ['{} {:1.2f}% ({})'.format(i, j, v) for i, j, v in
                    zip(NCBI_labels, 100. * values / values.sum(), values)]
```

```python
NCBI_main_labels
```

```python
NCBI_colors=palettable.colorbrewer.qualitative.Pastel1_9.mpl_colors
```

```python
plt.figure(figsize=(8,8))
plt.pie(NCBI_sizes,colors=NCBI_colors)
circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
plt.legend(NCBI_main_labels,loc='center right' ,bbox_to_anchor=(1.7, 0.5), fontsize=17,
           framealpha=1,fancybox=False,edgecolor='black')

plt.title('NCBI',fontsize=35,x=0.5,y=0.47) 
plt.savefig('Fig1.B.pdf', bbox_inches='tight')
plt.show()
```
