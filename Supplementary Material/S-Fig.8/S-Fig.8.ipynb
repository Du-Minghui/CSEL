```python
import pandas as pd
```

```python
bacteria = pd.read_csv("bacteria_strains.csv")
```

```python
bacteria
```

```python
bacteria_strain =list(set(bacteria["NP1"]).union(bacteria["NP2"],
                                                 bacteria["NP3"],
                                                 bacteria["NP4"],
                                                 bacteria["NP5"],
                                                 bacteria["NP6.1"],
                                                 bacteria["NP6.2"]))
```

```python
bacteria_strain
```

```python
heatmap = pd.DataFrame(index=bacteria_strain)
```

```python
heatmap.index.name ="bacteria strain"
```

```python
heatmap
```

```python
np = pd.read_csv("NP6.2.csv",index_col=0)
```

```python
np
```

```python
heatmap = heatmap.join(np["Confidence"])
heatmap = heatmap.rename(columns={'Confidence': 'NP6.2'})
heatmap 
```

```python
heatmap = heatmap.drop(heatmap.index[1])
```

```python
heatmap
```

```python
heatmap.to_csv("heatmap.csv",index_label = None)
```

```python
heatmap = heatmap.fillna(0)
```

```python
heatmap1=heatmap.sort_values(by=['NP1','NP2','NP3','NP4','NP5','NP6.1','NP6.2'],ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import palettable
```

```python
cmap=plt.get_cmap('YlGn')
```

```python
plt.figure(figsize=(20,50))
fig = sns.heatmap(data=heatmap1,vmin=0,vmax=1,cmap=cmap)
plt.savefig('heatmap.pdf', bbox_inches='tight') 
plt.show()
```
