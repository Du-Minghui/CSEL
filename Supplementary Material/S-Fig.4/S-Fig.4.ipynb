```python
import sqlite3, pandas as pd, numpy as np, string, matplotlib.pyplot as plt, seaborn as sns
```

```python
df = pd.read_csv("external_test.csv")
```

```python
df
```

```python
metrics = df["metrics"].tolist()
y1 = df["similarity_100%"].tolist()
y2 = df["similarity_95%"].tolist()
y3 = df["similarity_90%"].tolist()
y4 = df["similarity_85%"].tolist()
y5 = df["similarity_80%"].tolist()
y6 = df["similarity_75%"].tolist()
```

```python
x =np.arange(len(metrics))
bar_width = 0.1

plt.figure(figsize=(15,7))

bar1 = plt.bar(x-bar_width*2, y1, bar_width, align="center", color="#fd798f", label="similarity=100%(sample:6610)", alpha=0.5)
bar2 = plt.bar(x-bar_width, y2, bar_width, color="#98FB98", align="center", label="similarity95%(sample:6936)", alpha=0.5)
bar3 = plt.bar(x, y3, bar_width, color="#bdf6fe", align="center", label="similarity90%(sample:7432)", alpha=0.5)
bar4 = plt.bar(x+bar_width, y4, bar_width, color="#ffe5ad", align="center", label="similarity85%(sample:7864)", alpha=0.5)
bar5 = plt.bar(x+bar_width*2, y5, bar_width, color="#ffbacd", align="center", label="similarity80%(sample:8842)", alpha=0.5)
bar6 = plt.bar(x+bar_width*3, y6, bar_width, color="#ffa756", align="center", label="similarity75%(sample:9328)", alpha=0.5)

for bar in bar1:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+15,f'{height}%',ha='center')
for bar in bar2:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+12,f'{height}%',ha='center')
for bar in bar3:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+10,f'{height}%',ha='center')
for bar in bar4:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+7,f'{height}%',ha='center')    
for bar in bar5:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+5,f'{height}%',ha='center')    
for bar in bar6:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+1,f'{height}%',ha='center')    

plt.ylim(bottom=0,top=120)
plt.xlabel("Metrics")

plt.xticks(x+bar_width/2, metrics)
plt.legend()

plt.savefig("S-Fig.4.pdf")
plt.show()
```
