```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
from statannotations.Annotator import Annotator
```

```python
plt.rcParams["font.family"] = "Arial"
plt.rcParams['axes.unicode_minus'] = False  
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

```python
data = pd.read_csv("HPO_ACC.csv")
```

```python
data
```

```python
models = data["models"].tolist()
y1 = data["Optuna"].tolist()
y2 = data["Hyperopt"].tolist()
y3 = data["RandomizedSearch"].tolist()
y4 = data["unHPO"].tolist()
```

```python
x =np.arange(len(models))
bar_width = 0.2

plt.figure(figsize=(10,6))

bar1 = plt.bar(x-bar_width, y1, bar_width, align="center", color="#fc824a", label="Optuna", alpha=0.3)
bar2 = plt.bar(x, y2, bar_width, color="#ff028d", align="center", label="Hyperopt", alpha=0.5)
bar3 = plt.bar(x+bar_width, y3, bar_width, color="#1E90FF", align="center", label="RandomizedSearch", alpha=0.5)
bar4 = plt.bar(x+bar_width*2, y4, bar_width, color="#98FB98", align="center", label="unHPO", alpha=0.5)

for bar in bar1:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+2.5, f'{height}%',ha='center')
for bar in bar2:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+2.5, f'{height}%',ha='center')
for bar in bar3:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+2.5, f'{height}%',ha='center')
for bar in bar4:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height+2.5, f'{height}%',ha='center')   

plt.ylim(bottom=50,top=85)
#plt.xlabel("Metrics")

plt.xticks(x+bar_width/2, models)
plt.legend()

plt.savefig("Supplementary Fig.x.pdf")
plt.show()
```
